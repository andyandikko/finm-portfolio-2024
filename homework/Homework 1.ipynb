### 1. 
There are thousands of individual risky assets in which HMC can invest.  Explain why MV optimization across 1,000 securities is infeasible.


The MV solution requires us to invert the covariance matrix: 
. The term is highly unstable for large number of securities, giving also a unstable result. With 1,000 securities, the estimation error of means, volatilities, and covariances produce impactful errors in allocation. The latter happens due to the high sensitivity to changes in input data. If the input data in the out-of-sample is slightly different from the in-sample, the out-of-sample optimal portfolio will have considerably different weights from the in-sample.

Optimizing for a smaller number of assets reduces the allocation consequence of estimation errors in the moments of the assets.

1. Feasibility of MV Optimization Across 1,000 Securities
MV optimization across 1,000 securities is impractical due to computational challenges and the minuscule weight each security would need to hold in the portfolio. Optimizing such a large number involves an extensive number of variables (weights) and an enormous set of covariances to consider, which increases computational complexity exponentially. Additionally, the practical aspect of acquiring such fractional shares of securities where weights approach near zero becomes unfeasible in real market conditions, due to minimum trade sizes and transaction costs.

2. Conditions for Effective Two-Stage Optimization
In two-stage optimization, the first stage involves optimizing within individual asset classes, and the second stage involves optimizing across these classes. For this approach to effectively approximate a full MV optimization, the securities within each asset class must exhibit higher intraclass correlations compared to inter-class correlations. This low inter-class correlation ensures that each asset class behaves almost independently, making local optimizations within asset classes a good proxy for part of a global solution. If the simplex algorithm or another linear programming method is used in optimization, it tends to select corner solutions where some weights can be zero, making the effective partitioning critical for achieving near-optimal global results.

3. Classification of TIPS as a Separate Asset Class
Considering TIPS' unique characteristics, especially their inflation protection capability and distinct risk-return profile indicated by a Sharpe ratio greater than 1, they should form their own asset class. Their low correlation with other asset classes further justifies this separation, allowing for more targeted inflation risk management within portfolio optimizations. This distinct classification aids in precise control over exposure to inflation-related risks, which is not typically achievable when grouped with other fixed-income securities.

4. Focus on Real Returns in Portfolio Allocation
Focusing on real returns, which account for inflation, is critical because nominal returns can misrepresent the true economic value of portfolio gains by not considering the erosion of purchasing power. Real returns provide a more accurate basis for comparing the performance of different asset classes under varying inflationary conditions. The use of real returns affects the mean-variance optimization by altering the expected returns and covariance matrix used in the calculations, potentially leading to different optimal solutions than those based on nominal returns. The impact of inflation can particularly skew the risk assessments and correlations between assets, thus real returns offer a more stable foundation for MV optimization.

5. Adjusting the Optimization Problem to Reflect Bounded Solutions
To model the constraints used by Harvard, as depicted in Exhibits 5 and 6, the optimization problem should incorporate bounded inequality constraints for each asset class, limiting allocations within specified ranges (e.g., 0-100%). Additional constraints might specify narrower ranges or special conditions like those for cash (e.g., -50% to 100%). These modifications ensure the optimizer adheres to policy limits and practical investment guidelines, avoiding extreme allocations that could lead to undesirable risk exposures or liquidity issues. Integrating such constraints might involve using the simplex algorithm which optimizes at vertices of the feasible region, typically resulting in allocations being pushed to their bounds.

6. Zero Allocation to Domestic Equities and Bonds
The zero allocation to domestic equities and bonds might indicate that these asset classes, under the constraints imposed, do not adequately contribute to the desired portfolio outcomes across the target return spectrum. This outcome can be attributed to the simplex algorithm, which optimizes at the extreme points of the feasible region, effectively pushing some asset weights to zero when they do not sufficiently enhance the portfolio's risk-return profile.
```python
## Summary Statistics
from scipy.stats import describe

def summary_stats(data):

    data_mean = describe(data,axis = 0)[2] * 12
    data_sample_var = describe(data,axis = 0)[3] 
    data_sample_stdev = describe(data,axis = 0)[3] **0.5 * 12**0.5
    data_sharpe =  data_mean / data_sample_stdev 
    # print(f"Securities names: {data.columns}")
    # print(f"Mean excess returns of each securities: {data_mean}")
    # print(f"Var of securities: {data_sample_var}")
    # print(f"Standard deviations of securities: {data_sample_stdev}")
    # print(f"Sharpe ratio for each security: {data_sharpe}")

    result_1 = pd.DataFrame({"Asset": data.columns,
                            "Mean returns": data_mean,
                            "Stddev":data_sample_stdev,
                            "Sharpe": data_sharpe})

    # data validation
    mean_check = data.apply(lambda x: x.mean(), axis=0).to_numpy()

    stdev_check = data.apply(lambda x: np.std(x, ddof=1), axis=0).to_numpy()

    val1 = np.isclose(mean_check*12,data_mean,rtol=0.01).all()
    val2 = np.isclose(stdev_check*12**0.5,data_sample_stdev,rtol=0.01).all()
    if val1 and val2: 
        print(f"Security with best sharpe ratio: {data.columns[np.argmax(data_sharpe)]}")
    else:
        print("Error in calculations")
    return result_1

result_1 = summary_stats(data)
display(result_1)
```

```python
def find_corr(correlation_matrix,type: str = "min"):    
    

    if type == "min":
        matrix = np.absolute(correlation_matrix.to_numpy())
        val = matrix.min()
        corr = np.where(matrix == val)
    elif type=="max":
        matrix = correlation_matrix.to_numpy() - np.eye(correlation_matrix.shape[0])
        val =matrix.max()
        corr = np.where(matrix == val)
    else:
        raise ValueError
    row = corr[0][0]
    col = corr[0][1]
    row_name = correlation_matrix.index[row]
    col_name =  correlation_matrix.index[col]
    print(f"{type.capitalize()} absolute correlation is between '{row_name}' and '{col_name}' with a coefficient of {val:.7f}")
    return row_name, col_name, val

find_corr(correlation_matrix, "min")
find_corr(correlation_matrix, "max")



```

```python
## The MV frontier
from numpy import linalg

def tangency_portfolio(data: pd.DataFrame):
    """
    Calculate the weights of the tangency portfolio using mean excess returns and a returns DataFrame.
    `data` should be a DataFrame where rows are observations and columns are different assets.
    scipy.stats.describe must be present
    """
    data_mean = describe(data,axis = 0)[2] * 12
    securities = data.columns
    mean_excess_returns = np.reshape(data_mean, (-1,))
    covariance_matrix = data.cov(ddof=1).to_numpy()
    inverse_cov_matrix = linalg.inv(covariance_matrix)
    tangency_portfolio = (inverse_cov_matrix @ mean_excess_returns) / (np.ones(shape=(1, inverse_cov_matrix.shape[0])) @ inverse_cov_matrix @ mean_excess_returns)
    
    if not (round(tangency_portfolio.sum()) == 1):
        raise ValueError("The sum of the tangency portfolio weights does not equal 1.")
    weights_df = pd.DataFrame({"Asset": securities,
                               "Weights":tangency_portfolio})
    return weights_df

display(tangency_portfolio(data=data))
```

```python

portfolio_mean_excess_returns_df = tangency_portfolio(data=data)
merged_df =pd.merge(left=result_1,right=portfolio_mean_excess_returns_df,on="Asset", how="outer")
tangency_portfolio_annualised_mean = merged_df.eval("`Mean returns` * Weights").sum()
covariance_matrix = data.cov(ddof=1).to_numpy()
tangency_portfolio_variance = merged_df["Weights"].to_numpy().T @ covariance_matrix @  merged_df["Weights"].to_numpy()
tangency_portfolio_annualised_stdev = tangency_portfolio_variance**0.5 * 12**0.5
tangency_portfolio_sharpe = tangency_portfolio_annualised_mean/tangency_portfolio_annualised_stdev
print(f"Annualised mean returns for tangency portfolio = {tangency_portfolio_annualised_mean}")
print(f"Annualised volatility for tangency portfolio = {tangency_portfolio_annualised_stdev}")
print(f"Annualised Sharpe for tangency portfolio = {tangency_portfolio_sharpe}")
```

* Continue with the same data file as the previous section.

* Suppose the investor has a targeted mean excess return (per month) of $\mutarg$ = 0.01.

Build the following portfolios:

#### Equally-weighted (EW)
Rescale the entire weighting vector to have target mean $\mutarg$. Thus, the $i$ element of the weight vector is,
$$\wEW_i = \frac{1}{n}$$

#### “Risk-parity” (RP)
Risk-parity is a term used in a variety of ways, but here we have in mind setting the weight of the portfolio to be proportional to the inverse of its full-sample variance estimate. Thus, the $i$ element of the weight vector is,
$$\wRP_i = \frac{1}{\sigma_i^2}$$

#### Regularized (REG)
Much like the Mean-Variance portfolio, set the weights proportional to 
$$\wREG \sim \widehat{\Sigma}^{-1}\mux$$
but this time, use a regularized covariance matrix,
$$\widehat{\Sigma} = \frac{\Sigma + \Sigma_D}{2}$$
where $\Sigma_D$ denotes a *diagonal* matrix of the security variances, with zeros in the off-diagonals.

Thus, $\widehat{\Sigma}$ is obtained from the usual covariance matrix, $\Sigma$, but shrinking all the covariances to half their estimated values. 


### Comparing

In order to compare all these allocation methods, (those above, along with the tangency portfolio obtained in the previous section,) rescale each weight vector, such that it has targeted mean return of $\mutarg$.

* Calculate the performance of each of these portfolios over the sample.
* Report their mean, volatility, and Sharpe ratio. 
* How do these compare across the four allocation methods?

```python
target = 0.12
Comparison_dict = {}
weights = tangency_portfolio(data=data)
data_mean = describe(data,axis = 0)[2] * 12
expected_returns = np.reshape(data_mean, (-1,))
data_cov_matrix = data.cov(ddof=1).to_numpy()
inverse_cov_matrix = linalg.inv(data_cov_matrix)
allocation = ((np.ones(shape=(1, inverse_cov_matrix.shape[0])) @ inverse_cov_matrix @ expected_returns)) / ((expected_returns.T @ inverse_cov_matrix @ expected_returns)) * target
weights = (weights.iloc[:,1].to_numpy() * allocation)

portfolio_returns = weights @ expected_returns
annualised_portfolio_variance = weights @ data_cov_matrix @ weights.T * 12
annualised_standard_deviation = annualised_portfolio_variance**0.5
Comparison_dict["MV"]={"mean_annualised_return":portfolio_returns,
                                    "annualised_volatility":annualised_standard_deviation,
                                    "Sharpe ratio":portfolio_returns/annualised_standard_deviation}

print(f"MV portfolio have mean annualised returns of {portfolio_returns}")
print(f"MV portfolio have an annualised volatility of {annualised_standard_deviation}")
print(f"MV portfolio have Sharpe ratio of {portfolio_returns/annualised_standard_deviation}")
```

```python
# Equally Weighted portfolio
target = 0.12
securities = len(data.columns)
data_cov_matrix = data.cov().to_numpy()
weights = np.full(securities, 1/securities) 
# portfolio_mean = weights.T @ data[]
expected_returns = data.mean(axis=0).to_numpy()
portfolio_returns = weights @ expected_returns * 12
#scaling
weights = (target/portfolio_returns) * weights
portfolio_returns = weights @ expected_returns * 12
annualised_portfolio_variance = weights @ data_cov_matrix @ weights.T * 12
annualised_standard_deviation = annualised_portfolio_variance**0.5
Comparison_dict["equally_weighted"]={"mean_annualised_return":portfolio_returns,
                                    "annualised_volatility":annualised_standard_deviation,
                                    "Sharpe ratio":portfolio_returns/annualised_standard_deviation}

print(f"Equally weighted portfolio have mean annualised returns of {portfolio_returns}")
print(f"Equally weighted portfolio have an annualised volatility of {annualised_standard_deviation}")
print(f"Equally weighted portfolio have Sharpe ratio of {portfolio_returns/annualised_standard_deviation}")
```

```python
# Risk Parity
weights_risk_parity = (1 / (data_cov_matrix.diagonal()*12)) / (1/(data_cov_matrix.diagonal()*12)).sum()

expected_returns = data.mean(axis=0).to_numpy()
portfolio_returns = weights_risk_parity @ expected_returns * 12
weights_risk_parity = (target/portfolio_returns) * weights_risk_parity
portfolio_returns = weights_risk_parity @ expected_returns * 12
annualised_portfolio_variance =weights_risk_parity @ data_cov_matrix @ weights_risk_parity.T * 12
annualised_standard_deviation = annualised_portfolio_variance**0.5
Comparison_dict["risk_parity"]={"mean_annualised_return":portfolio_returns,
                                    "annualised_volatility":annualised_standard_deviation,
                                    "Sharpe ratio":portfolio_returns/annualised_standard_deviation}
print(f"Risk Parity portfolio have mean annualised returns of {portfolio_returns}")
print(f"Risk Parity portfolio have an annualised volatility of {annualised_standard_deviation}")
print(f"Risk Parity portfolio have Sharpe ratio of {portfolio_returns/annualised_standard_deviation}")
```

```python
variance_matrix = np.zeros_like(data_cov_matrix)  
np.fill_diagonal(variance_matrix, np.diag(data_cov_matrix)) 
regularized_cov_matrix = (variance_matrix + data_cov_matrix)/2
expected_returns = data.mean(axis=0).to_numpy()
weights = linalg.inv(regularized_cov_matrix) @ expected_returns
portfolio_returns = weights @ expected_returns * 12
#scaling
weights = (target/portfolio_returns) * weights
portfolio_returns = weights @ expected_returns * 12
annualised_portfolio_variance = weights @ data_cov_matrix @ weights.T * 12
annualised_standard_deviation = annualised_portfolio_variance**0.5
Comparison_dict["regularized"]={"mean_annualised_return":portfolio_returns,
                                    "annualised_volatility":annualised_standard_deviation,
                                    "Sharpe ratio":portfolio_returns/annualised_standard_deviation}

print(f"regularized portfolio have mean annualised returns of {portfolio_returns}")
print(f"regularized portfolio have an annualised volatility of {annualised_standard_deviation}")
print(f"regularized portfolio have Sharpe ratio of {portfolio_returns/annualised_standard_deviation}")
```

## 1. One-step Out-of-Sample (OOS) Performance
Let’s divide the sample to both compute a portfolio and then check its performance out of sample.
* Using only data through the end of `2022`, compute the weights built in Section 3.
* Rescale the weights, (using just the in-sample data,) to set each allocation to have the same mean return of $\mutarg$.
* Using those weights, calculate the portfolio’s Sharpe ratio within that sample.
* Again using those weights, (derived using data through `2022`,) calculate the portfolio’s OOS Sharpe ratio, which is based only on performance in `2023-2024`.

```python
filtered_data = data.loc["2011-02-28":"2022-12-31",:]

annualised_expected_returns = filtered_data.mean(axis = 0).to_numpy()
cov_matrix = filtered_data.cov().to_numpy()
inverse_cov_matrix = linalg.inv(cov_matrix)
    
def portfolio_statistics(data: pd.DataFrame, portfolio: str = "MV", target_return: float = 0.01, weights: np.array = None):
        
    annualised_expected_returns = data.mean(axis = 0).to_numpy() * 12
    cov_matrix = data.cov().to_numpy() * 12 # annualised cov matrix
    inverse_cov_matrix = linalg.inv(cov_matrix)
    
    def scale_weights(weights, target_return = target_return,annualised_expected_returns=annualised_expected_returns):
        portfolio_return = weights.T @ annualised_expected_returns
        scaling_factor = target_return / portfolio_return
        weights = scaling_factor * weights
        portfolio_return = weights.T @ annualised_expected_returns

        if abs(portfolio_return - target_return) <= target_return*0.001:
            return weights
        else:
            raise ValueError("Issues with scaling")
    if weights is not None:
         pass
    elif portfolio == "MV":
        ones = np.ones_like(annualised_expected_returns).T
        tangency_portfolio = inverse_cov_matrix @  annualised_expected_returns / (ones@inverse_cov_matrix@annualised_expected_returns)
        weights = ((ones @ inverse_cov_matrix @ annualised_expected_returns) / (annualised_expected_returns.T @ inverse_cov_matrix @ annualised_expected_returns)) * target_return * tangency_portfolio
    elif portfolio == "EW":
        weights = 1/(len(data.columns)) * np.ones_like(annualised_expected_returns)
        weights = scale_weights(weights=weights)
    
    elif portfolio == "RP":
        individual_var = cov_matrix.diagonal()
        weights = 1/individual_var
        weights = scale_weights(weights=weights)
        
    elif portfolio == "Reg":
        variance_matrix = np.zeros_like(cov_matrix)  
        np.fill_diagonal(variance_matrix, np.diag(cov_matrix)) 
        corrected_cov = (variance_matrix + cov_matrix)/2
        weights = corrected_cov @ annualised_expected_returns
        weights = scale_weights(weights=weights)
    else:
         raise NameError("Unrecognised portfolio optimisation symbol")
    portfolio_return = weights.T @ annualised_expected_returns
    portfolio_variance = weights.T @ cov_matrix @ weights
    portfolio_sharpe = portfolio_return / portfolio_variance**0.5
    return ({"Portfolio": portfolio,
                "Annualised_return":portfolio_return,
                "Annualised volatility": portfolio_variance**0.5,
                "Sharpe": portfolio_sharpe},weights)



MV_weights = portfolio_statistics(data = filtered_data,
                     portfolio="MV",
                     target_return=0.01)
EW_weights = portfolio_statistics(data = filtered_data,
                     portfolio="EW",
                     target_return=0.01)
RP_weights = portfolio_statistics(data = filtered_data,
                     portfolio="RP",
                     target_return=0.01)
Reg_weights = portfolio_statistics(data = filtered_data,
                     portfolio="Reg",
                     target_return=0.01)
```
