## Question 1.2

Long-Term Capital Management (LTCM) held significant advantages over competitors due to its expertise, quantitative methodologies, and strategic relationships. Led by John Meriwether and a team of renowned quantitative experts and Nobel laureates, LTCM leveraged unparalleled intellectual capital to develop proprietary models that allowed it to exploit subtle mispricings and systematically manage risk. These models, grounded in advanced mathematical and statistical theories, formed the backbone of LTCM’s strategy and gave the firm a significant edge in financial markets.

Proprietary Models and Applications
Prepayment Risk in Mortgage-Backed Securities (MBS):

LTCM developed a sophisticated model to analyze prepayment risks, which arise when homeowners refinance their mortgages during periods of falling interest rates.
Their model estimated the fair value of MBS securities and their sensitivity to interest rate changes, enabling LTCM to capitalize on mispricings overlooked by less sophisticated investors. For example, their analysis allowed them to predict how interest rate volatility would influence refinancing behavior, adjusting their positions accordingly.
Stock Price Dynamics:

LTCM employed stochastic models to predict stock price movements and measure exposure to market risks. By using methods like the Black-Scholes option pricing framework, they dynamically adjusted their positions in derivatives to exploit volatility mispricings while minimizing directional exposure to the underlying assets.
Interest Rate Models:

For fixed-income arbitrage, LTCM used term structure models, such as the Heath-Jarrow-Morton framework, to predict changes in the yield curve and identify relative value trades. These models allowed LTCM to take advantage of anomalies, such as the "butterfly spread" trades, where mid-term rates deviated from short-term and long-term rates.
Swap Spread Trades:

LTCM employed models to assess the fair value of swap spreads. For example, in Japanese Government Bonds (JGBs), their models evaluated spreads between bond yields and swap fixed rates while incorporating funding costs and market liquidity conditions. This enabled LTCM to predict mean-reversion and profit from misaligned spreads.
Volatility Models:

In its volatility-selling strategy, LTCM used models to compare implied volatility (used to price options) with realized volatility in the markets. For instance, when implied volatility of equity index options exceeded historical volatility, their models indicated a high probability of profits by selling options and dynamically hedging them.
Institutional Demand-Driven Mispricings:

LTCM modeled pricing anomalies created by institutional constraints such as regulatory requirements or operational inefficiencies. For example:
They exploited the pricing differential between Royal Dutch Petroleum and Shell, which traded at different valuations despite being identical in economic terms.
LTCM profited from market imbalances caused by banks and governments hedging large positions, such as in the European bond markets during the transition to the Euro.
Strategic Strengths
Low Financing Costs:

LTCM negotiated favorable financing terms, including low repo rates and long-term funding agreements. This efficiency allowed them to scale trades with minimal cost, amplifying returns on positions with small mispricings. For example, LTCM financed U.S. Treasury positions at LIBOR - 20 basis points, reducing carry costs and increasing net profitability.
Scalable Risk-Taking:

LTCM’s ability to use leverage at a ratio of 19:1 amplified the impact of small pricing discrepancies. For instance, a 1-basis-point change in swap spreads on a $1 billion notional trade generated $100,000 in profit, and LTCM’s portfolio was structured to maximize these incremental gains at scale.
Uncorrelated Strategies:

LTCM’s diversified portfolio spanned multiple asset classes and geographic regions. By incorporating uncorrelated risks, such as MBS prepayment risk, European bond convergence trades, and equity arbitrage, LTCM reduced overall portfolio volatility while maintaining high returns.
Local Market Insights:

LTCM partnered with institutional investors globally, gaining access to local perspectives on macroeconomic policies and market behaviors. These partnerships informed their strategies in regional markets, such as Japanese swap spreads or European government bonds, ensuring well-informed trades with minimized risk.
Quantitative Portfolio Construction:

LTCM applied advanced portfolio optimization techniques, balancing risk and return while ensuring zero beta exposure to broad market risks (e.g., equities, bonds, and currencies). This approach focused on maximizing returns from idiosyncratic mispricings rather than relying on directional market movements.

## Question 1.3

Management:

```python

def load_data(data_path: str, filename: str, sheet_name: str):
    full_data_path = os.path.join(data_path, filename)
    data = pd.read_excel(full_data_path, sheet_name=sheet_name, parse_dates=True, skiprows=2, index_col=0)
    data = data.dropna()
    if "date" in data.columns:
        data = data.set_index("date")
    elif "Date" in data.columns:
        data = data.set_index("Date")
    return data

# Load LTCM and GMO data
data_path = "./data"
data_2 = load_data(data_path, "ltcm_exhibits_data.xlsx", sheet_name="Exhibit 2")
data_4 = load_data(data_path, "ltcm_exhibits_data.xlsx", sheet_name="Exhibit 4")

def load_gmo_data(data_path: str, filename: str, sheet_name: str):
    full_data_path = os.path.join(data_path, filename)
    data = pd.read_excel(full_data_path, sheet_name=sheet_name, parse_dates=True, index_col=0)
    data = data.dropna()
    return data

spy_data = load_gmo_data(data_path, "gmo_analysis_data.xlsx", sheet_name="total returns")
rf_data = load_gmo_data(data_path, "gmo_analysis_data.xlsx", sheet_name="risk-free rate")

# Align indices and convert to period-based monthly data
data_2.index = pd.to_datetime(data_2.index).to_period("M")
data_4.index = pd.to_datetime(data_4.index).to_period("M")
spy_data.index = pd.to_datetime(spy_data.index).to_period("M")
rf_data.index = pd.to_datetime(rf_data.index).to_period("M")

# Adjust risk-free rate for alignment
rf_data["TBill 3M"] = rf_data["TBill 3M"].shift(1)
spy_data["SPY"] = spy_data["SPY"].shift(1)
# Merge SPY and Risk-Free data
spy_excess_data = pd.merge(spy_data, rf_data, left_index=True, right_index=True, how="left").dropna()
spy_excess_data["SPY_Excess"] = spy_excess_data["SPY"] - spy_excess_data["TBill 3M"]

# Merge LTCM data with Risk-Free data
data_2_excess = pd.merge(data_2, rf_data, left_index=True, right_index=True, how="left").dropna()
data_2_excess["excess_gross"] = data_2_excess["Gross Monthly Performancea"] - data_2_excess["TBill 3M"]
data_2_excess["excess_net"] = data_2_excess["Net Monthly Performanceb"] - data_2_excess["TBill 3M"]
data_2_excess = data_2_excess.loc[:, ["excess_gross", "excess_net"]]

# Function to calculate statistics
def calculate_stats(data: pd.DataFrame, annual_factor: int = 12, quantile: float = 0.05):
    if isinstance(data, pd.Series):
        data = data.to_frame()
    stats = pd.DataFrame(index=data.columns)
    stats["skew"] = data.skew()
    stats["kurtosis"] = data.kurtosis()
    stats["mean"] = data.mean()
    stats["std"] = data.std()
    stats["sharpe"] = stats["mean"] / stats["std"]
    stats["annualized_mean"] = stats["mean"] * annual_factor
    stats["annualized_std"] = stats["std"] * np.sqrt(annual_factor)
    stats["annualized_sharpe"] = stats["sharpe"] * np.sqrt(annual_factor)
    stats["VaR"] = data.quantile(quantile)
    stats["CVaR"] = data[data < data.quantile(quantile)].mean()
    drawdown = data.apply(lambda x: (1 + x).cumprod() / (1 + x).cumprod().cummax() - 1)
    stats["max_drawdown"] = drawdown.min()
    return stats

# Calculate statistics for LTCM and SPY data
ltcma_stats = calculate_stats(data_2_excess)
spy_stats = calculate_stats(spy_excess_data.loc[data_2_excess.first_valid_index():data_2_excess.last_valid_index(),"SPY_Excess"])

# Display results
display(ltcma_stats)
display(spy_stats)
```

The static carry trade involves borrowing in a low-interest-rate currency and investing in a high-interest-rate currency, assuming a consistent positive return based solely on the interest rate differential. However, this approach does not account for fluctuations in exchange rates, which can erode or reverse these returns. By incorporating conditional risk premia, the carry trade can be significantly improved.

The conditional approach leverages forecasts for the FX risk premium to determine when the carry trade is likely to generate positive returns. Specifically, trades are only executed when the expected FX risk premium is positive:This ensures that the trade not only benefits from favorable interest rate differentials but also aligns with predicted favorable exchange rate movements. For instance, if the interest rate differential suggests a profitable trade but the regression model predicts a negative risk premium due to anticipated adverse exchange rate movements, the trade can be avoided. This dynamic filtering reduces the likelihood of engaging in loss-making trades.

The improved strategy dynamically adjusts based on market conditions, effectively timing the carry trade to maximize profitability and reduce risk. By avoiding trades when the risk premium is negative, the strategy avoids scenarios where exchange rate depreciation offsets interest rate gains. Furthermore, the conditional approach can help mitigate risks during periods of heightened market volatility or safe-haven flows, where traditional carry trades often suffer losses. Over time, this results in higher average returns and improved risk-adjusted performance compared to the static carry trade.
