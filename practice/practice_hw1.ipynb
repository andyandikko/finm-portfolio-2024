### 1. 
There are thousands of individual risky assets in which HMC can invest.  Explain why MV optimization across 1,000 securities is infeasible.


The MV solution requires us to invert the covariance matrix: 
. The term is highly unstable for large number of securities, giving also a unstable result. With 1,000 securities, the estimation error of means, volatilities, and covariances produce impactful errors in allocation. The latter happens due to the high sensitivity to changes in input data. If the input data in the out-of-sample is slightly different from the in-sample, the out-of-sample optimal portfolio will have considerably different weights from the in-sample.

Optimizing for a smaller number of assets reduces the allocation consequence of estimation errors in the moments of the assets.

1. Feasibility of MV Optimization Across 1,000 Securities
MV optimization across 1,000 securities is impractical due to computational challenges and the minuscule weight each security would need to hold in the portfolio. Optimizing such a large number involves an extensive number of variables (weights) and an enormous set of covariances to consider, which increases computational complexity exponentially. Additionally, the practical aspect of acquiring such fractional shares of securities where weights approach near zero becomes unfeasible in real market conditions, due to minimum trade sizes and transaction costs.

2. Conditions for Effective Two-Stage Optimization
In two-stage optimization, the first stage involves optimizing within individual asset classes, and the second stage involves optimizing across these classes. For this approach to effectively approximate a full MV optimization, the securities within each asset class must exhibit higher intraclass correlations compared to inter-class correlations. This low inter-class correlation ensures that each asset class behaves almost independently, making local optimizations within asset classes a good proxy for part of a global solution. If the simplex algorithm or another linear programming method is used in optimization, it tends to select corner solutions where some weights can be zero, making the effective partitioning critical for achieving near-optimal global results.

3. Classification of TIPS as a Separate Asset Class
Considering TIPS' unique characteristics, especially their inflation protection capability and distinct risk-return profile indicated by a Sharpe ratio greater than 1, they should form their own asset class. Their low correlation with other asset classes further justifies this separation, allowing for more targeted inflation risk management within portfolio optimizations. This distinct classification aids in precise control over exposure to inflation-related risks, which is not typically achievable when grouped with other fixed-income securities.

4. Focus on Real Returns in Portfolio Allocation
Focusing on real returns, which account for inflation, is critical because nominal returns can misrepresent the true economic value of portfolio gains by not considering the erosion of purchasing power. Real returns provide a more accurate basis for comparing the performance of different asset classes under varying inflationary conditions. The use of real returns affects the mean-variance optimization by altering the expected returns and covariance matrix used in the calculations, potentially leading to different optimal solutions than those based on nominal returns. The impact of inflation can particularly skew the risk assessments and correlations between assets, thus real returns offer a more stable foundation for MV optimization.

5. Adjusting the Optimization Problem to Reflect Bounded Solutions
To model the constraints used by Harvard, as depicted in Exhibits 5 and 6, the optimization problem should incorporate bounded inequality constraints for each asset class, limiting allocations within specified ranges (e.g., 0-100%). Additional constraints might specify narrower ranges or special conditions like those for cash (e.g., -50% to 100%). These modifications ensure the optimizer adheres to policy limits and practical investment guidelines, avoiding extreme allocations that could lead to undesirable risk exposures or liquidity issues. Integrating such constraints might involve using the simplex algorithm which optimizes at vertices of the feasible region, typically resulting in allocations being pushed to their bounds.

6. Zero Allocation to Domestic Equities and Bonds
The zero allocation to domestic equities and bonds might indicate that these asset classes, under the constraints imposed, do not adequately contribute to the desired portfolio outcomes across the target return spectrum. This outcome can be attributed to the simplex algorithm, which optimizes at the extreme points of the feasible region, effectively pushing some asset weights to zero when they do not sufficiently enhance the portfolio's risk-return profile. Conceptually, in an unconstrained optimization, there might be small allocations to these classes if they offered some diversification or return benefit without significantly increasing risk.

7. Impact of Tightened Constraints on the Mean-Variance Tradeoff
The tightening of allocation constraints as seen in Exhibit 6 leads to a noticeable deterioration in the mean-variance tradeoff, where the Sharpe ratio declines from 0.38 to about 0.35 or 0.36, reflecting a roughly 5% reduction in portfolio efficiency. This change indicates that tighter constraints restrict the optimizer's ability to allocate capital to optimally balance risk and return, thus reducing the overall effectiveness of the portfolio in achieving higher risk-adjusted returns. This effect is particularly pronounced when the optimizer is constrained from allocating to assets that might otherwise improve the portfolio's performance under different risk-return objectives.

```python
## The MV frontier
from numpy import linalg

def tangency_portfolio(data: pd.DataFrame):
    """
    Calculate the weights of the tangency portfolio using mean excess returns and a returns DataFrame.
    `data` should be a DataFrame where rows are observations and columns are different assets.
    scipy.stats.describe must be present
    """
    data_mean = describe(data,axis = 0)[2] * 12
    securities = data.columns
    mean_excess_returns = np.reshape(data_mean, (-1,))
    covariance_matrix = data.cov(ddof=1).to_numpy()
    inverse_cov_matrix = linalg.inv(covariance_matrix)
    tangency_portfolio = (inverse_cov_matrix @ mean_excess_returns) / (np.ones(shape=(1, inverse_cov_matrix.shape[0])) @ inverse_cov_matrix @ mean_excess_returns)
    
    if not (round(tangency_portfolio.sum()) == 1):
        raise ValueError("The sum of the tangency portfolio weights does not equal 1.")
    weights_df = pd.DataFrame({"Asset": securities,
                               "Weights":tangency_portfolio})
    return weights_df

display(tangency_portfolio(data=data))
```

Build the following portfolios:

#### Equally-weighted (EW)
Rescale the entire weighting vector to have target mean $\mutarg$. Thus, the $i$ element of the weight vector is,
$$\wEW_i = \frac{1}{n}$$

#### “Risk-parity” (RP)
Risk-parity is a term used in a variety of ways, but here we have in mind setting the weight of the portfolio to be proportional to the inverse of its full-sample variance estimate. Thus, the $i$ element of the weight vector is,
$$\wRP_i = \frac{1}{\sigma_i^2}$$

#### Regularized (REG)
Much like the Mean-Variance portfolio, set the weights proportional to 
$$\wREG \sim \widehat{\Sigma}^{-1}\mux$$
but this time, use a regularized covariance matrix,
$$\widehat{\Sigma} = \frac{\Sigma + \Sigma_D}{2}$$
where $\Sigma_D$ denotes a *diagonal* matrix of the security variances, with zeros in the off-diagonals.

Thus, $\widehat{\Sigma}$ is obtained from the usual covariance matrix, $\Sigma$, but shrinking all the covariances to half their estimated values.

## 1. One-step Out-of-Sample (OOS) Performance
Let’s divide the sample to both compute a portfolio and then check its performance out of sample.
* Using only data through the end of `2022`, compute the weights built in Section 3.
* Rescale the weights, (using just the in-sample data,) to set each allocation to have the same mean return of $\mutarg$.
* Using those weights, calculate the portfolio’s Sharpe ratio within that sample.
* Again using those weights, (derived using data through `2022`,) calculate the portfolio’s OOS Sharpe ratio, which is based only on performance in `2023-2024`.

```python
filtered_data = data.loc["2011-02-28":"2022-12-31",:]
import numpy as np
import pandas as pd

def portfolio_statistics(
    data: pd.DataFrame, 
    portfolio: str = "MV", 
    target_return: float = 0.01, 
    weights: np.ndarray = None
) -> tuple[dict[str, float], np.ndarray]:
    """
    Calculate portfolio statistics and optimal weights based on the specified optimization strategy.
    
    Parameters:
    ----------
    data : pd.DataFrame
        Historical return data of assets, with each column representing an asset and each row representing a time period.
        
    portfolio : str, optional
        The optimization strategy to use. Options include:
        - "MV" : Minimum Variance Portfolio
        - "EW" : Equally Weighted Portfolio
        - "RP" : Risk Parity Portfolio
        - "Reg" : Regularized Portfolio
        - "Tangency" : Tangency Portfolio
        Default is "MV".
    
    target_return : float, optional
        The desired annualized return for the portfolio. Used for scaling weights. Default is 0.01 (1%).
    
    weights : np.ndarray, optional
        Pre-defined portfolio weights. If None, weights will be calculated based on the chosen optimization strategy.
    
    Returns:
    -------
    tuple[dict[str, float], np.ndarray]
        A tuple containing:
        - A dictionary with portfolio statistics:
            * "Portfolio" : str (Type of portfolio optimization used)
            * "Annualised_return" : float (Expected annualized return)
            * "Annualised volatility" : float (Annualized standard deviation)
            * "Sharpe" : float (Sharpe ratio)
        - The computed optimal weights as a numpy array.
    
    Raises:
    ------
    ValueError
        If there are issues with scaling weights to match the target return.
    
    NameError
        If an unrecognized portfolio optimization symbol is provided.
    """
    
    # Calculate annualized expected returns and covariance matrix
    annualised_expected_returns = data.mean(axis=0).to_numpy() * 12
    cov_matrix = data.cov().to_numpy() * 12  # annualized covariance matrix
    inverse_cov_matrix = np.linalg.inv(cov_matrix)
    
    # Function to scale weights to match the target return
    def scale_weights(weights: np.ndarray, target_return: float = target_return) -> np.ndarray:
        portfolio_return = weights.T @ annualised_expected_returns
        scaling_factor = target_return / portfolio_return
        weights = scaling_factor * weights
        portfolio_return = weights.T @ annualised_expected_returns

        if abs(portfolio_return - target_return) <= target_return * 0.001:
            return weights
        else:
            raise ValueError("Issues with scaling")

    if weights is not None:
        pass
    elif portfolio == "MV":
        ones = np.ones_like(annualised_expected_returns).T
        tangency_portfolio = (
            inverse_cov_matrix @ annualised_expected_returns 
            / (ones @ inverse_cov_matrix @ annualised_expected_returns)
        )
        weights = (
            (ones @ inverse_cov_matrix @ annualised_expected_returns) 
            / (annualised_expected_returns.T @ inverse_cov_matrix @ annualised_expected_returns)
        ) * target_return * tangency_portfolio
    elif portfolio == "EW":
        weights = 1 / len(data.columns) * np.ones_like(annualised_expected_returns)
        weights = scale_weights(weights=weights)
    elif portfolio == "RP":
        individual_var = cov_matrix.diagonal()
        weights = 1 / individual_var
        weights = scale_weights(weights=weights)
    elif portfolio == "Reg":
        corrected_cov = (np.diag(np.diag(cov_matrix)) + cov_matrix) / 2
        weights = np.linalg.inv(corrected_cov) @ annualised_expected_returns
        weights = scale_weights(weights=weights)
    elif portfolio == "Tangency":
        weights = np.linalg.inv(cov_matrix) @ annualised_expected_returns
        weights = scale_weights(weights=weights)
    else:
        raise NameError("Unrecognised portfolio optimisation symbol")
    
    # Calculate portfolio statistics
    portfolio_return = weights.T @ annualised_expected_returns
    portfolio_variance = weights.T @ cov_matrix @ weights
    portfolio_sharpe = portfolio_return / portfolio_variance**0.5
    
    return (
        {
            "Portfolio": portfolio,
            "Annualised_return": portfolio_return,
            "Annualised volatility": portfolio_variance**0.5,
            "Sharpe": portfolio_sharpe
        }, 
        weights
    )

# Example usage of the function
MV_weights = portfolio_statistics(
    data=filtered_data, 
    portfolio="MV", 
    target_return=0.12
)
EW_weights = portfolio_statistics(
    data=filtered_data, 
    portfolio="EW", 
    target_return=0.12
)
RP_weights = portfolio_statistics(
    data=filtered_data, 
    portfolio="RP", 
    target_return=0.12
)
Reg_weights = portfolio_statistics(
    data=filtered_data, 
    portfolio="Reg", 
    target_return=0.12
)
Tangency_weights = portfolio_statistics(
    data=filtered_data, 
    portfolio="Tangency", 
    target_return=0.12
)
```

## 2. Rolling OOS Performance

Iterate the Out-of-Sample performance every year, not just the final year. Namely,
* Start at the end of `2015`, and calculate the weights through that time. Rescale them using the mean returns through that time.
* Apply the weights to the returns in the upcoming year, (`2016`.)
* Step forward a year in time, and recompute.
* Continue until again calculating the weights through `2023` and applying them to the returns in `2024`.

Report the mean, volatility, and Sharpe from this dynamic approach for the following portfolios:
* mean-variance (tangency)
* equally-weighted
* risk-parity
* regularized

```python
def rolling_oos_performance(
        assets_excess_returns,
        in_sample_end_date,
        out_of_sample_start_date,
        out_of_sample_end_date,
        MU_MONTH_TARGET=0.01
    ):
    assets_excess_returns = assets_excess_returns.copy()
    n_assets = assets_excess_returns.shape[1]
    in_sample_assets_excess_returns = assets_excess_returns.loc[:in_sample_end_date]
    out_of_sample_assets_excess_returns = (
        assets_excess_returns.loc[out_of_sample_start_date:out_of_sample_end_date]
    )

    # Regularized
    in_sample_weights_regularized = pmh.calc_tangency_weights(in_sample_assets_excess_returns, cov_mat=.5, name="Regularized")

    # Tangency
    in_sample_weights_tangency = pmh.calc_tangency_weights(in_sample_assets_excess_returns)

    # # No TIPS Tangency
    # in_sample_weights_no_tips = pmh.calc_tangency_weights(in_sample_assets_excess_returns_no_tips, name="No TIPS Tangency")

    # # Modified TIPS Tangency
    # in_sample_weights_modified_tips = pmh.calc_tangency_weights(in_sample_assets_excess_returns_modified_tips, name="Mod TIPS Tangency")

    # Risk Parity
    in_sample_asset_variance_dict = in_sample_assets_excess_returns.std().map(lambda x: x ** 2).to_dict()
    in_sample_asset_inv_variance_dict = {asset: 1 / variance for asset, variance in in_sample_asset_variance_dict.items()}
    in_sample_weights_risk_parity = pd.DataFrame(in_sample_asset_inv_variance_dict, index=["Risk Parity Weights"]).transpose()

    # Equal Weights
    in_sample_weights_equal = pd.DataFrame(
        data=[[1 / n_assets] for _ in range(n_assets)],
        columns=["Equal Weights"],
        index=in_sample_assets_excess_returns.columns
    )

    # Join Weights
    in_sample_weights = (
        pd.concat([
            in_sample_weights_regularized,
            in_sample_weights_tangency,
            # in_sample_weights_no_tips,
            # in_sample_weights_modified_tips,
            in_sample_weights_risk_parity,
            in_sample_weights_equal
        ], axis=1)
        .fillna(0)
    )

    # Rescale Weights
    in_sample_weights_scaled = (
        in_sample_weights
        .apply(lambda weights: weights * MU_MONTH_TARGET / (in_sample_assets_excess_returns @ weights).mean())
    )

    return out_of_sample_assets_excess_returns @ in_sample_weights_scaled


oos_portfolios_performance = pd.DataFrame({})

for in_sample_last_year in range(2015, 2024):
    oos_portfolios_yearly_performance = rolling_oos_performance(
        data,
        f"{in_sample_last_year}-12-31",
        f"{in_sample_last_year+1}-01-01",
        f"{in_sample_last_year+1}-12-31",
        MU_MONTH_TARGET=0.01
    )
    oos_portfolios_performance = pd.concat([oos_portfolios_performance, oos_portfolios_yearly_performance])

display(oos_portfolios_performance)

display(pmh.calc_cummulative_returns(oos_portfolios_performance))
display(pmh.calc_summary_statistics(
    oos_portfolios_performance,
    annual_factor=12,
    provided_excess_returns=True,
    keep_columns=['Annualized Mean', 'Annualized Vol', 'Annualized Sharpe']
))
```

```python
import numpy as np
import pandas as pd

def find_first_date(data: pd.DataFrame, month: int, year: int = None) -> pd.Timestamp:
    """
    Find the first date in the specified month and optional year.
    
    Parameters:
    ----------
    data : pd.DataFrame
        DataFrame with a DatetimeIndex.
        
    month : int
        Month to filter (1 for January, 12 for December, etc.).
        
    year : int, optional
        Year to filter. If None, the function will ignore the year.
    
    Returns:
    -------
    pd.Timestamp
        The first date matching the criteria.
    """
    if year:
        filtered_dates = data.loc[(data.index.month == month) & (data.index.year == year)]
    else:
        filtered_dates = data.loc[data.index.month == month]
        
    return filtered_dates.index.min() if not filtered_dates.empty else None

def calculate_portfolio_statistics(
    data: pd.DataFrame,
    target_monthly: float = 0.01
) -> pd.DataFrame:
    """
    Calculate weights and portfolio returns for different optimization strategies.
    
    Parameters:
    ----------
    data : pd.DataFrame
        Historical return data of assets.
        
    target_monthly : float, optional
        Desired monthly return to scale weights.
    
    Returns:
    -------
    pd.DataFrame
        Annualized mean, volatility, and Sharpe ratio for each portfolio type.
    """
    tangency_weights = np.linalg.inv(data.cov()) @ data.mean() / (np.ones(data.mean().shape) @ np.linalg.inv(data.cov()) @ data.mean())
    ew_weights = np.repeat(1 / data.columns.size, data.columns.size)
    rp_weights = 1 / data.var()
    reg_weights = np.linalg.inv((data.cov() + np.diag(np.diag(data.cov()))) / 2) @ data.mean()

    weights = np.array([tangency_weights, ew_weights, rp_weights, reg_weights])
    portfolio_weighted_returns_unscaled = data @ weights.T

    # Calculate unique scaling factors for each portfolio
    scaling_factors = target_monthly / portfolio_weighted_returns_unscaled.mean(axis=0)
    portfolio_weighted_scaled = weights * scaling_factors.to_numpy().reshape(-1, 1)

    # Calculate scaled portfolio returns
    portfolio_weighted_returns = data @ portfolio_weighted_scaled.T
    
    # Calculate portfolio statistics
    def calculate_statistics(returns: pd.DataFrame, annual_factor: int = 12) -> pd.DataFrame:
        mean = returns.mean() * annual_factor
        vol = returns.std() * np.sqrt(annual_factor)
        sharpe = mean / vol
        return pd.concat([mean, vol, sharpe], axis=1, keys=["Mean", "Vol", "Sharpe"])

    return calculate_statistics(portfolio_weighted_returns), portfolio_weighted_returns

def rolling_portfolio_statistics(
    data: pd.DataFrame,
    insample_start: str,
    insample_end: str,
    oos_start: str,
    oos_end: str,
    target_monthly: float = 0.01
) -> tuple[pd.DataFrame, pd.DataFrame, pd.DataFrame]:


    insample_data = data.loc[insample_start:insample_end]
    oos_data = data.loc[oos_start:oos_end]

    # In-sample statistics and returns
    insample_stats, insample_returns = calculate_portfolio_statistics(insample_data, target_monthly=target_monthly)
    
    # Calculate weights for out-of-sample based on in-sample weights
    tangency_weights = np.linalg.inv(insample_data.cov()) @ insample_data.mean() / (np.ones(insample_data.mean().shape) @ np.linalg.inv(insample_data.cov()) @ insample_data.mean())
    ew_weights = np.repeat(1 / insample_data.columns.size, insample_data.columns.size)
    rp_weights = 1 / insample_data.var()
    reg_weights = np.linalg.inv((insample_data.cov() + np.diag(np.diag(insample_data.cov()))) / 2) @ insample_data.mean()

    weights = np.array([tangency_weights, ew_weights, rp_weights, reg_weights])
    portfolio_weighted_returns_unscaled = oos_data @ weights.T

    # Scale weights and calculate OOS statistics
    scaling_factors = target_monthly / portfolio_weighted_returns_unscaled.mean(axis=0)
    portfolio_weighted_scaled = weights * scaling_factors.to_numpy().reshape(-1, 1)
    portfolio_weighted_returns = oos_data @ portfolio_weighted_scaled.T
    
    oos_stats, oos_returns = calculate_portfolio_statistics(portfolio_weighted_returns)
    
    return insample_returns, oos_returns, insample_stats, oos_stats


def iterate_across_years(data: pd.DataFrame, target_monthly: float = 0.01) -> list:
    """
    Iterate across all possible years in the data to calculate rolling statistics.
    
    Parameters:
    ----------
    data : pd.DataFrame
        Historical return data of assets.
        
    target_monthly : float, optional
        Desired monthly return to scale weights.
    
    Returns:
    -------
    list
        A list of tuples containing in-sample and out-of-sample statistics and returns for each rolling period.
    """
    results = []
    years = data.index.year.unique()
    
    for year in years[:-1]:  # Ensure there is a next year for OOS
        insample_start = f"{year}-01-01"
        insample_end = f"{year}-12-31"
        oos_start = f"{year + 1}-01-01"
        oos_end = f"{year + 1}-12-31"
        
        insample_returns, oos_returns, insample_stats, oos_stats = rolling_portfolio_statistics(
            data, insample_start, insample_end, oos_start, oos_end, target_monthly=target_monthly
        )
        
        results.append((insample_returns, oos_returns, insample_stats, oos_stats))
    
    return results

def iterate_across_years(data: pd.DataFrame, target_monthly: float = 0.01) -> pd.DataFrame:
    """
    Iterate across all possible years in the data to calculate rolling statistics.
    
    Parameters:
    ----------
    data : pd.DataFrame
        Historical return data of assets.
        
    target_monthly : float, optional
        Desired monthly return to scale weights.
    
    Returns:
    -------
    pd.DataFrame
        A concatenated DataFrame containing in-sample and out-of-sample returns across all rolling periods.
    """
    all_returns = []  # To store all returns across rolling periods
    years = data.index.year.unique()
    
    for year in years[:-1]:  # Ensure there is a next year for OOS
        insample_start = f"{year}-01-01"
        insample_end = f"{year}-12-31"
        oos_start = f"{year + 1}-01-01"
        oos_end = f"{year + 1}-12-31"
        
        insample_returns, oos_returns, _, _ = rolling_portfolio_statistics(
            data, insample_start, insample_end, oos_start, oos_end, target_monthly=target_monthly
        )
        
        # Add a year identifier to the returns DataFrames for tracking
        insample_returns["Period"] = f"{year} (In-Sample)"
        oos_returns["Period"] = f"{year + 1} (Out-of-Sample)"
        
        # Concatenate insample and oos returns for this year
        combined_returns = pd.concat([insample_returns, oos_returns])
        all_returns.append(combined_returns)
    
    # Concatenate all years' returns into a single DataFrame
    final_returns_df = pd.concat(all_returns)
    return final_returns_df
def calculate_statistics(returns: pd.DataFrame, annual_factor: int = 12) -> pd.DataFrame:
    """
    Calculate mean, volatility, and Sharpe ratio for given returns.
    
    Parameters:
    ----------
    returns : pd.DataFrame
        Portfolio returns for which to calculate statistics.
        
    annual_factor : int, optional
        Factor to annualize returns and volatility. Default is 12 (monthly data).
        
    Returns:
    -------
    pd.DataFrame
        DataFrame containing mean, volatility, and Sharpe ratio for each portfolio.
    """
    mean = returns.mean() * annual_factor
    vol = returns.std() * np.sqrt(annual_factor)
    sharpe = mean / vol
    return pd.concat([mean, vol, sharpe], axis=1, keys=["Mean", "Vol", "Sharpe"])

def group_and_calculate_statistics(all_rolling_returns: pd.DataFrame) -> pd.DataFrame:
    """
    Group the rolling returns by the 'Period' column and calculate statistics.
    
    Parameters:
    ----------
    all_rolling_returns : pd.DataFrame
        DataFrame containing portfolio returns across rolling periods.
    
    Returns:
    -------
    pd.DataFrame
        DataFrame of aggregated statistics for each period.
    """
    # Group by 'Period' and calculate annualized statistics for each group
    grouped_stats = all_rolling_returns.groupby("Period").apply(lambda x: calculate_statistics(x.iloc[:, :-1]))
    
    return grouped_stats

# Example usage:
all_rolling_returns = iterate_across_years(data)
grouped_stats = group_and_calculate_statistics(all_rolling_returns)

print("Annualized Portfolio Statistics Grouped by Period:")
display(grouped_stats)

# Example usage:
all_rolling_returns = iterate_across_years(data)
print("All Rolling Returns Across Periods:\n")
display(all_rolling_returns)
# # Reset the index so that 'Date' becomes a column
# all_rolling_returns_reset = all_rolling_returns.reset_index()

# # Filter using boolean conditions on the 'Date' column
# filtered_returns = all_rolling_returns_reset[
#     (all_rolling_returns_reset['Date'] >= pd.to_datetime("2016-01-31")) &
#     (all_rolling_returns_reset['Date'] <= pd.to_datetime("2016-12-31"))
# ]

# print("Filtered Returns for 2016:")
# display(filtered_returns)
# Reset the index so that 'Date' becomes a column
all_rolling_returns_reset = all_rolling_returns.reset_index()

# Rename the index column to 'Date' if not already done
all_rolling_returns_reset.rename(columns={"index": "Date"}, inplace=True)

# Convert the 'Date' column to datetime if needed
all_rolling_returns_reset["Date"] = pd.to_datetime(all_rolling_returns_reset["Date"])

# Define start and end dates
start = pd.to_datetime("2016-01-31")
end = pd.to_datetime("2024-12-31")

# Use boolean masking to filter based on the 'Date' column
filtered_returns = all_rolling_returns_reset[
    (all_rolling_returns_reset["Date"] >= start) &
    (all_rolling_returns_reset["Date"] <= end)
]

# Drop the 'Period' column before passing to the cumulative returns function
filtered_returns_clean = filtered_returns.drop(columns=["Period", "Date"])

# Example call to your function
pmh.calc_cummulative_returns(filtered_returns_clean)
```

```python

def portfolio_statistics(
    data: pd.DataFrame, 
    portfolio: str = "MV", 
    target_return: float = 0.01, 
    weights: np.ndarray = None
) -> tuple[dict[str, float], np.ndarray]:
    """
    Calculate portfolio statistics and optimal weights based on the specified optimization strategy.
    
    Parameters:
    ----------
    data : pd.DataFrame
        Historical return data of assets, with each column representing an asset and each row representing a time period.
        
    portfolio : str, optional
        The optimization strategy to use. Options include:
        - "MV" : Minimum Variance Portfolio
        - "EW" : Equally Weighted Portfolio
        - "RP" : Risk Parity Portfolio
        - "Reg" : Regularized Portfolio
        - "Tangency" : Tangency Portfolio
        Default is "MV".
    
    target_return : float, optional
        The desired annualized return for the portfolio. Used for scaling weights. Default is 0.01 (1%).
    
    weights : np.ndarray, optional
        Pre-defined portfolio weights. If None, weights will be calculated based on the chosen optimization strategy.
    
    Returns:
    -------
    tuple[dict[str, float], np.ndarray]
        A tuple containing:
        - A dictionary with portfolio statistics:
            * "Portfolio" : str (Type of portfolio optimization used)
            * "Annualised_return" : float (Expected annualized return)
            * "Annualised volatility" : float (Annualized standard deviation)
            * "Sharpe" : float (Sharpe ratio)
        - The computed optimal weights as a numpy array.
    
    Raises:
    ------
    ValueError
        If there are issues with scaling weights to match the target return.
    
    NameError
        If an unrecognized portfolio optimization symbol is provided.
    """
    
    # Calculate annualized expected returns and covariance matrix
    annualised_expected_returns = data.mean(axis=0).to_numpy() * 12
    cov_matrix = data.cov().to_numpy() * 12  # annualized covariance matrix
    inverse_cov_matrix = np.linalg.inv(cov_matrix)
    
    # Function to scale weights to match the target return
    def scale_weights(weights: np.ndarray, target_return: float = target_return) -> np.ndarray:
        portfolio_return = weights.T @ annualised_expected_returns
        scaling_factor = target_return / portfolio_return
        weights = scaling_factor * weights
        portfolio_return = weights.T @ annualised_expected_returns

        if abs(portfolio_return - target_return) <= target_return * 0.001:
            return weights
        else:
            raise ValueError("Issues with scaling")

    if weights is not None:
        pass
    elif portfolio == "MV":
        ones = np.ones_like(annualised_expected_returns).T
        tangency_portfolio = (
            inverse_cov_matrix @ annualised_expected_returns 
            / (ones @ inverse_cov_matrix @ annualised_expected_returns)
        )
        weights = (
            (ones @ inverse_cov_matrix @ annualised_expected_returns) 
            / (annualised_expected_returns.T @ inverse_cov_matrix @ annualised_expected_returns)
        ) * target_return * tangency_portfolio
    elif portfolio == "EW":
        weights = 1 / len(data.columns) * np.ones_like(annualised_expected_returns)
        weights = scale_weights(weights=weights)
    elif portfolio == "RP":
        individual_var = cov_matrix.diagonal()
        weights = 1 / individual_var
        weights = scale_weights(weights=weights)
    elif portfolio == "Reg":
        corrected_cov = (np.diag(np.diag(cov_matrix)) + cov_matrix) / 2
        weights = np.linalg.inv(corrected_cov) @ annualised_expected_returns
        weights = scale_weights(weights=weights)
    elif portfolio == "Tangency":
        weights = np.linalg.inv(cov_matrix) @ annualised_expected_returns
        weights = scale_weights(weights=weights)
    else:
        raise NameError("Unrecognised portfolio optimisation symbol")
    
    # Calculate portfolio statistics
    portfolio_return = weights.T @ annualised_expected_returns
    portfolio_variance = weights.T @ cov_matrix @ weights
    portfolio_sharpe = portfolio_return / portfolio_variance**0.5
    
    return (
        {
            "Portfolio": portfolio,
            "Annualised_return": portfolio_return,
            "Annualised volatility": portfolio_variance**0.5,
            "Sharpe": portfolio_sharpe
        }, 
        weights
    )

def time_series_split(data, train_start_date,test_length_months=12):
    train_start_date = pd.to_datetime(train_start_date)
    train_end_date = train_start_date + pd.DateOffset(months=test_length_months)
    test_end_date = train_end_date + pd.DateOffset(months=test_length_months)

    filtered_data = data.loc[train_start_date:train_end_date]
    data_OOS = data.loc[train_end_date + pd.DateOffset(days=1):test_end_date]

    return filtered_data, data_OOS

start_date = pd.to_datetime('2015-12-31')
last_date = pd.to_datetime('2024-09-30')
results = []
while start_date + pd.DateOffset(years=1) <= last_date:
    filtered_data, data_OOS = time_series_split(data=data, train_start_date=start_date,
                                                test_length_months=12)

    annualised_expected_returns = filtered_data.mean(axis = 0).to_numpy()
    cov_matrix = filtered_data.cov().to_numpy()
    inverse_cov_matrix = np.linalg.inv(cov_matrix)

    MV_weights = portfolio_statistics(data = filtered_data,
                        portfolio="MV",
                        target_return=0.01)
    EW_weights = portfolio_statistics(data = filtered_data,
                        portfolio="EW",
                        target_return=0.01)
    RP_weights = portfolio_statistics(data = filtered_data,
                        portfolio="RP",
                        target_return=0.01)
    Reg_weights = portfolio_statistics(data = filtered_data,
                        portfolio="Reg",
                        target_return=0.01)

    OOS = [portfolio_statistics(data=data_OOS, portfolio= i[0]["Portfolio"],weights = i[1]) for i in [MV_weights, EW_weights, RP_weights, Reg_weights]]

    print(f"OOS results from training on {str(start_date.date())} to {str((start_date + pd.DateOffset(years=1)).date())}")
    display(pd.DataFrame([i[0] for i in OOS]))
    start_date = start_date + pd.DateOffset(years=1)
```
